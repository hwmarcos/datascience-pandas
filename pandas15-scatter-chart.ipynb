```python
iris = pydataset.data('iris')
```

```python
iris.head()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], sizes=iris['Petal.Length']*50)
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2
```

```python
iris['specie_number'] = iris['Species'].apply(specie_color)
```

```python
iris.head()
```

```python
plt.scatter(
    iris['Sepal.Length'], iris['Sepal.Width'], 
    sizes=iris['Petal.Length']*100, 
    c=iris['specie_number'],
    cmap='viridis',
    alpha=0.5
)
```
