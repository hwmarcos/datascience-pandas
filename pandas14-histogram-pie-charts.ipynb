```python
estados = pd.read_csv('populacao_brasil.csv')
```

```python
estados.sort_values(['total'], ascending=False) #order by total asc
```

```python
estados['total'].hist() #grafico de histograma (distribuio de frequncia)
```

```python
fig, ax = plt.subplots()
plt.hist(estados['total'], bins=15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['percent'] = (estados['total'] / estados['total'].sum()) * 100
```

```python
estados.head()
```

```python
estados['percent'].plot(kind='pie')
```

```python
plt.pie(estados['percent'], labels=estados['estado'], autopct='%1.2f%%')
plt.show()
```
